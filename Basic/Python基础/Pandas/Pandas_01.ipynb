```python
# 排序
df.sort_values(by='Q1') # 按Q1列数据升序排序
```

```python
df.sort_values(by='Q1',ascending=False) # 按Q1列数据降序排序
```

```python
# 分组聚合
df.groupby('team').sum() # 按team分组对应列相加
```

```python
df.groupby('team').mean()
```

```python
# 不同列不同的计算方式
df.groupby('team').agg({'Q1':sum,
                       'Q2':'count',
                       'Q3':'mean',
                       'Q4':max})
```

```python
# 数据转换
df.groupby('team').sum().T
```

```python
df.groupby('team').sum().stack()
```

```python
df.groupby('team').sum().unstack()
```

```python
# 增加列
df['one'] = 1
df
```

```python
df['total']=df.Q1+df.Q2+df.Q3+df.Q4
df
```

```python
df['total']=df.loc[:,'Q1':'Q4'].apply(lambda x:sum(x), axis=1)
df
```

```python
df['total']=df.sum(axis=1)
df
```

```python
df['avg']=df.total/4
df
```

```python
# 统计分析
df.mean() # 返回所有列的均值
df.mean(1) # 返回所有行的均值，下同
df.corr() # 返回列与列之间的相关系数
df.count() # 返回每一列中的非空值的个数
df.max() # 返回每一列的最大值
df.min() # 返回每一列的最小值
df.median() # 返回每一列的中位数
df.std() # 返回每一列的标准差
df.var() # 方差
df.mode() # 众数
```
